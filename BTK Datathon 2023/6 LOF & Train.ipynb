## Verimizi içeri aktaralım

```python
# Veriyi içeri aktaralım.

# veri yolumuzu verelim
file_path = 'train2.csv'

# verimizi okuyoruz. 
df = pd.read_csv(file_path)

print("Setup Complete")
```

## Bağımlı değişkenimizi Belirleyelim

```python
# target verilerimiz string olduğu için LabelEncoder yapıyoruz. Her sınıfı bir sayıya atıyoruz.
l_fit = LabelEncoder().fit(df["Öbek İsmi"])
y = l_fit.transform(df["Öbek İsmi"])
#y = pd.DataFrame(y, columns = ["Type_1"]) # dataframe'e çevirmek için bu satırı kullanabiliriz.
y[0:20]
```

```python
# labellerimizi buluyoruz.
y_label = l_fit.inverse_transform([i for i in range(max(y)+1)])
label_and_num = list(zip(y_label, [i for i in range(max(y)+1)]))
label_and_num
```

```python
df["Öbek İsmi"] = y
```

## Object Kolonlarımızı Buluyoruz

```python
object_cols, d = calc_object_cols(df, 0)
d
```

```python
# Object kolonlarımıza OrdinalEncoder uyguluyoruz. OneHot denemesi yapılacak!
ordinal_encoder = OrdinalEncoder()
df[object_cols] = ordinal_encoder.fit_transform(df[object_cols])
df
```

## Outlier Verilerimizi Saptayalım ve Eğitim Verimizeden Çıkaralım

```python
# Local Outlier
lof = LocalOutlierFactor()
lof.fit(df)
pred = lof.fit_predict(df)
np.unique(pred, return_counts=True)
```

```python
df["inlier"] = pred
```

```python
# Outlier olan verilerimizi filtreleyelim.
rslt_df = df[df['inlier'] == -1]
```

```python
# Filtrelediğimiz outlier verilerimizin indekslerini alalım.
rslt_df.index
```

```python
# DataFrame'mimizin içinden outlier verilerimizi drop edelim.
df = df.drop(index=rslt_df.index)
df.shape
```
